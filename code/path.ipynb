```python
import matplotlib.pyplot as plt

# 経路の定義（コーナーを原点とする）
path_x = [2.3, 0, 0]  # X座標
path_y = [0, 0, 3.3]  # Y座標


source_x,source_y = ([0.5, 1.5, 2.0, -0.3, 0.3, -0.3, 0.3, -0.3, -0.3],
 [0.3, 0.3, -0.3, 0.5, 1.0, 1.5, 2.0, 2.5, -0.3])


anomaly_x,anomaly_y = ([1.,],[-0.3])



# 経路のプロット
plt.figure(figsize=(6, 8))
plt.plot(path_x, path_y, label='Robot Path', linewidth=2)  # Plot the path

# 音源のプロット
plt.scatter(source_x, source_y, color='red', label='Source Position')  # Plot the sources

# グラフ設定
plt.title('Robot Path and Source Positions')
plt.xlabel('X Axis (m)')
plt.ylabel('Y Axis (m)')
plt.axis('equal')
plt.legend()

# グラフの表示
plt.show()
```

```python
import matplotlib.pyplot as plt

# 経路の定義（コーナーを原点とする）
path_x = [2.3, 0, 0]  # X coordinates
path_y = [0, 0, 3.3]  # Y coordinates

# 音源の位置
source_x, source_y = ([0.5, 1.5, 2.0, -0.3, 0.3, -0.3, 0.3, -0.3, -0.3],
                      [0.3, 0.3, -0.3, 0.5, 1.0, 1.5, 2.0, 2.5, -0.3])

# 異常位置
anomaly_x, anomaly_y = ([1.], [-0.3])

# 経路のプロット
plt.figure(figsize=(6, 8))
plt.plot(path_x, path_y, label='Robot Path', linewidth=2)  # Plot the path

# 音源のプロット
plt.scatter(source_x, source_y, color='red', label='Source Position')  # Plot the sources

# 異常点のプロット
plt.scatter(anomaly_x, anomaly_y, color='blue', marker='*', s=100, label='Anomaly')  # Plot the anomalies

# グラフ設定
plt.title('Robot Path and Source Positions')
plt.xlabel('X Axis (m)')
plt.ylabel('Y Axis (m)')
plt.axis('equal')
plt.legend()

# グラフの表示
plt.show()
```
